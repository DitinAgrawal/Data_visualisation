```python
data = data[-data.Rating.isnull()]
data.isna().sum()
```

```python
data = data[-data['Content Rating'].isnull()]
data.isna().sum()
```

```python
data = data[-data['Current Ver'].isnull()]
data.isna().sum()
```

```python
data = data[-data['Android Ver'].isnull()]
data.isna().sum()
```

```python
data
```

```python
data.columns
```

```python
data.Price = data.Price.apply(lambda x: 0 if x =='0' else float(x[1:]))
data['Price'].dtype
```

```python
data.Reviews =  data.Reviews.astype("int32")
```

```python
data['Reviews'].dtype
```

```python
data.info()
```

```python
data.Size = data.Size.apply(lambda x: 0 if x =='0' or x=='Varies with device' else float(x[:len(x)-2]))
```

```python
data.Size
```

```python
data.Installs = data.Installs.apply(lambda x: 0 if x =='0' else x[:len(x)-2])
```

```python
sns.scatterplot(x=data['Price'],y=data['Rating'],hue=data['Size'])
```

```python
data=data[data.Price<=50]
```

```python
sns.scatterplot(x=data['Price'],y=data['Rating'],hue=data['Size'])
```

```python
sns.scatterplot(x=data['Price'],y=data['Installs'],hue=data['Size'])
```
